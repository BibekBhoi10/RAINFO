```python
import pandas as pd
df = pd.read_csv('sample.csv')
df.head()
```

```python
X = df.drop(df[["PrecipitationTomorrow","RainTomorrow"]], axis = 'columns')
y_class = df.RainTomorrow
y_reg = df.PrecipitationTomorrow
```

```python
from sklearn.model_selection import train_test_split

# Split the full dataset into training and testing (80% train, 20% test)
X_train, X_test, y_class_train, y_class_test, y_reg_train, y_reg_test = train_test_split(
    X, y_class, y_reg, test_size=0.2, random_state=42
)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor

rf_class = RandomForestClassifier()
rf_class.fit(X_train, y_class_train)  # predict 0 or 1

# Filter rows where rain actually happened
X_rain = X_train[y_class_train == 1]
y_rain_amount = y_reg_train[y_class_train == 1]

rf_reg = RandomForestRegressor()
rf_reg.fit(X_rain, y_rain_amount)

# Predict rain (yes or no)
rain_pred = rf_class.predict(X_test)

# Predict amount (only for predicted rain == 1)
rain_amount_pred = []

for i in range(len(X_test)):
    if rain_pred[i] == 1:
        amt = rf_reg.predict([X_test.iloc[i]])[0]
    else:
        amt = 0.0
    rain_amount_pred.append(amt)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_class_test, rain_pred)
print("Accuracy:", accuracy)

from sklearn.metrics import r2_score

r2 = r2_score(y_reg_test, rain_amount_pred)
print("R2 Score:", r2)
```
